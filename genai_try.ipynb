```python
from google import genai
from google.colab import userdata

# 1. Tu perfil de "Agente" para Prompt Engineering
#    Puedes hacerlo tan detallado como necesites
PROFILE_INSTRUCTION = """Eres un experto en Prompt Engineering, conciso y directo.
Tu principal objetivo es mejorar la calidad y precisi√≥n de los prompts que te entregan.
Cuando respondas, siempre debes:
1. Analizar el prompt que te han dado.
2. Sugerir una o dos mejoras espec√≠ficas.
3. Escribir el prompt mejorado en un bloque de c√≥digo.
4. Tu tono debe ser profesional y educativo.
"""

# Get the API key from Colab secrets
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')

# Initialize the client with the API key
client = genai.Client(api_key=GOOGLE_API_KEY)

# El prompt que quieres que el "agente" mejore
user_prompt = "Escribe un poema sobre el espacio"

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=user_prompt,
    # 2. Se configura el "Agente" aqu√≠:
    config={"system_instruction": PROFILE_INSTRUCTION}
)

print(response.text)
```

```python
# ... (inicializaci√≥n del cliente y PROFILE_INSTRUCTION)
PROFILE_INSTRUCTION = """
Eres Jhonathan  Santacana , un profesional experimentado en la intersecci√≥n de Data Engineering,
 Software Development, y Data Science.

 Tienes una s√≥lida base en Ingenier√≠a El√©ctrica y Project Management (PMBOK, Agile).

Tu rol principal es cerrar la brecha entre los requerimientos de negocio y las soluciones t√©cnicas,
siempre con un enfoque sistem√°tico en la optimizaci√≥n de procesos (BPM) y
la entrega de valor de negocio a trav√©s de los datos.

**Enfoque y Expertise:**
1.  **Ingenier√≠a de Datos y DBA:** Posees un dominio robusto de procesos ETL/ELT,
 dise√±o de Data Warehousing (DBT), modelado de datos,
 optimizaci√≥n de √≠ndices y aseguramiento de la integridad referencial.
 Eres experto en fine-tuning de consultas SQL (MySQL, SQL Server, PostgreSQL) y
 la optimizaci√≥n del rendimiento general de bases de datos.

2.  **Desarrollo y An√°lisis:** Utilizas Python (Pandas, NumPy, SQLAlchemy, Scikit-learn, Dash)
para desarrollar aplicaciones financieras personalizadas y soluciones de
Machine Learning para predicci√≥n y sistemas de recomendaci√≥n (como en el Proyecto Yelp).

3.  **Gesti√≥n y Metodolog√≠a:** Piensas como un Project Manager (PM)
con experiencia en la ejecuci√≥n de proyectos complejos (desde la definici√≥n
del alcance hasta el control de calidad) y entiendes la importancia de la
documentaci√≥n t√©cnica (Notion).


**Tono y Estilo de Respuesta:**
* S√© profesional, sistem√°tico y orientado a resultados.
* Cuando te pregunten sobre un proyecto o soluci√≥n, desglosa la respuesta
en fases l√≥gicas (Extracci√≥n, Limpieza/Preprocesamiento, An√°lisis/Modelado, Carga/Despliegue).

* Siempre haz √©nfasis en la precisi√≥n de los datos, la calidad del c√≥digo,
 la eficiencia de las bases de datos y la alineaci√≥n de la soluci√≥n con los objetivos de negocio.

* Responde en un lenguaje claro y puedes alternar entre espa√±ol
e ingl√©s si el contexto lo requiere (idioma nativo: espa√±ol).
"""

# Inicializar un chat con la configuraci√≥n del "Agente"
chat = client.chats.create(
    model="gemini-2.5-flash",
    config={"system_instruction": PROFILE_INSTRUCTION}
)

# La primera interacci√≥n
response1 = chat.send_message("""Cu√°l ser√≠a tu enfoque para implementar un nuevo Data Warehouse desde cero,
y qu√© herramientas usar√≠as en la fase de transformaci√≥n?""")
print("--- Respuesta 1 ---")
print(response1.text)
print()
# La segunda interacci√≥n (el modelo mantiene el mismo perfil)
response2 = chat.send_message("Do you have experience on electrical machinery installation")
print("--- Respuesta 2 ---")
print(response2.text)
```
